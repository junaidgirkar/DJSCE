```python
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://en.wikipedia.org/wiki/Dwarkadas_J._Sanghvi_College_of_Engineering"
page = requests.get(URL)

soup = BeautifulSoup(page.content, "html.parser")
```

```python
URL2 = "https://quotes.toscrape.com/"
page2 = requests.get(URL2)
soup2 = BeautifulSoup(page2.content, "html.parser")
```

```python
# Find element by ID
results2 = soup2.find_all(class_="quote")
print(results2)
```

```python
for quote in results2:
  quote = quote.find(class_="text")
  print(quote.text.strip())
```

```python
# Find element by ID
results = soup.find(id="content")
print(results.prettify())
```

```python
# Find elements by tag and class name
page_titles = results.find_all("span", class_="mw-headline")
print("Page Titles:\n", page_titles)
```

```python
for title in page_titles:
    print(title, end="\n"*2)
```

```python
# Extract text from HTML elements
for title in page_titles:
    print(title.text.strip())
```

```python
# Extracting Main class containing departments
results2 = soup.find(class_="mw-parser-output")
results2.prettify()
```

```python
departments = results2.find_all("a", class_="mw-redirect")
departments
```

```python
for department in departments:
  print(department.text)
```
